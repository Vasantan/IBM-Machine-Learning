```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model_rmse_comparison = pd.DataFrame({'Model': ['NMF', 'NN Embedding','Linear Regression', 'Lasso Regression'], 'RMSE': [1.3088, 0.1202, 0.8138,0.8138]})
```

```python

palette = sns.color_palette('husl', len(model_rmse_comparison))  # 'husl' for different colors
sns.barplot(data=model_rmse_comparison, x='Model', y='RMSE', palette=palette)

plt.title('RMSE Comparison')
plt.ylabel('RMSE')
plt.xticks(rotation=45)
plt.xlabel('Model')

# Add values on top of bars
for index, row in model_rmse_comparison.iterrows():
    plt.text(index, row['RMSE'], f"{row['RMSE']:.2f}", ha='center', va='bottom')

plt.show()
```
